```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_theme(style='darkgrid', color_codes=True)
```

```python
sns.relplot(x='Date', y='Confirmed', data=covid_country_wise_data.iloc[::10, :], kind='line', height=4.5, aspect=2)
plt.ylabel('Rate', fontsize=14)
plt.xlabel('Date', fontsize=14)
plt.title('Daily Confirmed COVID-19 Cases Globally', fontsize=16)
plt.xticks(rotation=90)
```

### Pie chart vs. donut

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'flour', 'suger', 'butter', 'berries'
sizes = [375, 75, 250, 300]
explode = (0, 0.0, 0, 0)  # only "explode" the 2nd slice 

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Standard: A pie in baker', size=14)
plt.show()
```

```python
fig, ax = plt.subplots()
import numpy as np

size = 0.3
ax.pie(sizes, radius=1, labels=labels,autopct='%1.1f%%',startangle=90,
       wedgeprops=dict(width=size, edgecolor='w'))

ax.set(aspect="equal")
plt.title('Donut: A pie in baker', size=14)
plt.show()
```

### Build radar chart using Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
categories = ['Food Quality', 'Food Variety', 
              'Service Quality', 'Ambiance', 'Affordability']
restaurant_1 = [4, 4, 3, 4, 3]
restaurant_2 = [5, 5, 4, 5, 2]
restaurant_3 = [3, 4, 5, 3, 5]
```

```python
label_loc = np.linspace(start=0, stop=2*np.pi,num=len(restaurant_1))
```

```python
plt.figure(figsize=(8,8))
plt.subplot(polar=True)
plt.plot(label_loc, restaurant_1, label='restaurant 1')
plt.plot(label_loc, restaurant_2, label='restaurant 2')
plt.plot(label_loc, restaurant_3, label='restaurant 3')
plt.title('Restaurant comparison', size=20)
lines, labels=plt.thetagrids(np.degrees(label_loc),labels=categories)
plt.legend()
plt.show()
```

observation: The last data point isnt connected to the first one
to solve: add an additional element to both categories and restaurants thats identical to the first item

```python
categories = ['Food Quality', 'Food Variety', 
              'Service Quality', 'Ambiance', 'Affordability']
categories = [*categories, categories[0]]

restaurant_1 = [4, 4, 3, 4, 3]
restaurant_2 = [5, 5, 4, 5, 2]
restaurant_3 = [3, 4, 5, 3, 5]

restaurant_1 = [*restaurant_1, restaurant_1[0]]
restaurant_2 = [*restaurant_2, restaurant_2[0]]
restaurant_3 = [*restaurant_3, restaurant_3[0]]
```

```python
label_loc = np.linspace(start=0, stop=2*np.pi,num=len(restaurant_1))

plt.figure(figsize=(8,8))
plt.subplot(polar=True)
plt.plot(label_loc, restaurant_1, label='restaurant 1')
plt.plot(label_loc, restaurant_2, label='restaurant 2')
plt.plot(label_loc, restaurant_3, label='restaurant 3')
plt.title('Restaurant comparison', size=20)
lines, labels=plt.thetagrids(np.degrees(label_loc),labels=categories)
plt.legend()
plt.show()
```

### Waterfall chart using matplotlib

```python
import numpy as np
import matplotlib

def waterfall(df, x, y):

    # calculate running totals
    df['tot'] = df[y].cumsum()
    df['tot1']=df['tot'].shift(1).fillna(0)

    # lower and upper points for the bar charts
    lower = df[['tot','tot1']].min(axis=1)
    upper = df[['tot','tot1']].max(axis=1)

    # mid-point for label position
    mid = (lower + upper)/2

    # positive number shows green, negative number shows red
    df.loc[df[y] >= 0, 'color'] = 'green'
    df.loc[df[y] < 0, 'color'] = 'red'

    # calculate connection points
    connect= df['tot1'].repeat(3).shift(-1)
    connect[1::3] = np.nan

    #fig,ax = plt.subplots()
    plt.figure()
    plt.figure(facecolor='gray') 
    ax = plt.axes()

    # plot first bar with colors
    bars = ax.bar(x=df[x],height=upper, color =df['color'])

    # plot second bar - invisible
    plt.bar(x=df[x], height=lower,color='white')

    # plot connectors
    plt.plot(connect.index,connect.values, 'k' )
    ax.set_facecolor('white')
    plt.title("Category-wise varied sales", size = 14)

    # plot bar labels
    for i, v in enumerate(upper):
        plt.text(i-.15, mid[i], f"{df[y][i]:,.0f}")

df = pd.DataFrame({'category':['Sales','Service','Expenses','Taxes','Interest'],
                   'num':[100,10,-20,-30,60]})
waterfall(df,'category','num')
```
